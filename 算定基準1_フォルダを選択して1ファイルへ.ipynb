全店舗のファイルが入ったフォルダを1枚のシートにまとめる

```python
import pandas as pd
import glob
import os
from google.colab import files
import zipfile

print("全店舗ファイルのフォルダを選択してください")

# 1. 複数のCSVファイルをアップロードする
uploaded = files.upload()
if not uploaded:
    print("ファイルがアップロードされませんでした。")
else:
    file_name_upload = list(uploaded.keys())[0]

    # 2. ZIPファイルの場合、展開する
    if file_name_upload.endswith('.zip'):
        with zipfile.ZipFile(file_name_upload, 'r') as zip_ref:
            zip_ref.extractall('csv_data')
        target_dir = 'csv_data'
    else:
        target_dir = '.'

    # 3. CSVファイルを読み込んでデータをまとめる
    combined_data = []
    # 読み込み対象のファイルをリストアップ
    csv_files = glob.glob(f'{target_dir}/**/*.csv', recursive=True)

    for file in csv_files:
        try:
            # Shift-JIS(cp932)で読み込み
            df = pd.read_csv(file, encoding='cp932')
            if df.empty:
                continue

            # 最終行を取得
            total_row = df.iloc[-1]

            # 各値の取得（列番号は0から始まるため注意）
            j_val    = total_row.iloc[8]   # 9列目
            s_val    = total_row.iloc[17]  # 18列目
            u_val    = total_row.iloc[19]  # 20列目
            w_val    = total_row.iloc[21]  # 22列目
            ce_val   = total_row.iloc[81]  # 82列目
            cg_val   = total_row.iloc[83]  # 84列目
            ci_val   = total_row.iloc[85]  # 86列目
            bg_val   = total_row.iloc[57]  # 58列目
            cc_val   = total_row.iloc[79]  # 80列目
            as_val   = total_row.iloc[43]  # 44列目
            csum_val = ce_val + cg_val + ci_val

            folder_name = os.path.basename(os.path.dirname(file))
            # filenameから「店」を削除
            clean_filename = os.path.basename(file).replace('店', '')

            # データをリストに追加
            combined_data.append({
                'Folder': folder_name,
                'filename': clean_filename,
                'J_稼働数': j_val,
                'S_算定区分1': s_val,
                'U_算定区分2': u_val,
                'W_算定区分3': w_val,
                'ce+cg+ci_延長支援': csum_val,
                'BG_専門実施': bg_val,
                'CC_送迎加算': cc_val,
                'AS_欠席加算': as_val
            })
        except Exception as e:
            print(f"エラー（スキップします）: {file} - {e}")

    # 4. DataFrameに変換して保存
    if combined_data:
        result_df = pd.DataFrame(combined_data)

        # 出力ファイル名を定義
        output_file = '複数を1ファイルに.csv'

        # 保存（ここを修正しました）
        result_df.to_csv(output_file, index=False, encoding='cp932')

        print(f"処理が完了しました。'{output_file}' をダウンロードします。")
        # ダウンロード（ここを修正しました）
        files.download(output_file)
    else:
        print("CSVファイルからデータが見つかりませんでした。")
```
